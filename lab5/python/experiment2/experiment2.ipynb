```python
# Generating test inputs
import os

for i in range(100, 15000, 100):
    os.system("python kp_generate.py 20000" + " " + str(i) + " " + "2000 " + str(i) + ".txt")
```

```python
# The generated file is used as input to the algorithm and timed
import time
import os
import pandas as pd

result = pd.DataFrame(columns=['knapsack_capacity', 'time'])

for i in range(100, 15000, 100):
    print("test: " + str(i) + ".txt")
    start = time.time()
    os.system("python dp_kp.py " + str(i) + ".txt")
    end = time.time()
    time_count = end - start
    print("time: " + str(time_count))
    result.loc[i] = [i, time_count]
```

```python
result.to_csv("test_result_2.csv")
result
```
