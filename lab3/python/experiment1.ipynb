```python
from hashset import hashset
import random
import time
import pandas as pd

def random_value(size = 7):
    return ''.join(random.sample(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'], size))
df = pd.DataFrame(columns=['set_size', 'average_insert', 'average_find', 'collision_rate', 'time'])
```

```python
for i in range(500, 100000, 100):
    print(i)
    words = []
    for j in range(0, i):
        words.append(random_value())
    test_set = hashset(i*2)
    start = time.time()
    for item in words:
        test_set.insert(item)
    for item in words:
        test_set.find(item)
    end = time.time()
    df.loc[df.index.size] = [i, test_set.averageInsert(), test_set.averageFind(), test_set.collisionRate(), end-start]
```

```python
print(len(df))
df.to_csv("hashset_result.csv")
df
```
