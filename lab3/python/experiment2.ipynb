```python
from bstree import bstree
import random
import time
import pandas as pd

def random_value(size = 7):
    return ''.join(random.sample(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'], size))
df = pd.DataFrame(columns=['set_size', 'average_insert', 'average_find', 'time'])
df2 = pd.DataFrame(columns=['set_size', 'average_insert', 'average_find', 'time'])
```

```python
for i in range(500, 100000, 200):
    print(i)
    words = []
    for j in range(0, i):
        words.append(random_value())
    test_bstree = bstree()
    start = time.time()
    for item in words:
        test_bstree.insert(item)
    for item in words:
        test_bstree.find(item)
    end = time.time()
    df.loc[df.index.size] = [i, test_bstree.findNum()/test_bstree.size(), test_bstree.findNum()/test_bstree.size(), end-start]
```

```python
print(len(df))
df
```

```python
df.plot.scatter(x='set_size', y='time')
df.plot.scatter(x='set_size', y='average_insert')
df.plot.scatter(x='set_size', y='average_find')
```
